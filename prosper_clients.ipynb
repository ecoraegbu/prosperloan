```python
columns_to_drop = ['listingkey', 'listingnumber', 'listingcreationdate', 'closeddate', 'occupation', 
                   'employmentstatusduration', 'currentlyingroup', 'datecreditpulled', 'firstrecordedcreditline', 
                   'totalcreditlinespast7years',  'openrevolvingaccounts', 'openrevolvingmonthlypayment', 
                   'inquirieslast6months', 'totalinquiries', 'incomeverifiable', 'incomerange', 'ontimeprosperpayments', 
                   'totalprosperpaymentsbilled', 'prosperpaymentslessthanonemonthlate', 'prosperpaymentsonemonthpluslate', 'investmentfromfriendsamount', 
                   'investors', 'recommendations', 'percentfunded', 'loanoriginationquarter',
                   'loanoriginationdate', 'loanoriginalamount', 'scorexchangeattimeoflisting',]
```

```python
def drop_columns(dataframe, columns_to_drop):
    for columns in columns_to_drop:
        try:
            i = columns_to_drop.index(columns)
            #print(i)
            dataframe.drop(columns = columns, inplace = True)
        except Exception as e:
            print(i, e) 
    return 
```

```python
drop_columns(prosper, columns_to_drop)
```
